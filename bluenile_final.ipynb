```python
# -* coding: utf-8 *-
#author: dtlin
#data: 2020-04-19
#descriptinon: 爬取bluenile的详细信息，并将爬取的数据存储到CSV文
 
import requests
from bs4 import BeautifulSoup
import pandas as pd
from fake_useragent import UserAgent 
import re

URL1="https://www.bluenile.com/cn/diamonds/round-cut"

name="bluenile"
 
class bluenileSpider(object):
 
    def __init__(self):
        self.headers = {"User-Agent": UserAgent().random}
        self.datas = list()
 

 
 
    def parsePage(self, url):
        #maxPage = self.getMaxPage(url)
        
        
        print("当前正在爬取: {}".format(url))
        
        response = requests.get(url, headers = self.headers)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.text, "html.parser")        
        a= soup.find_all("a", role = "row")
            #print(a)
        num = 0
        for k in a:
            c = k.get("href")
                #print(c)
            url2 = "http://www.bluenile.com/hk/chi"
            url = url2 + c[2:]
            
            detail = self.parseDetail(url)
            num = num + 1
            print("Had download" + str(num) + "page")
            self.datas.append(detail)
        
       
 
       
        data = pd.DataFrame(self.datas)
        # columns字段：自定义列的顺序（DataFrame默认按列名的字典序排序）
        columns = ["序号","价格", "成色", "净度", "克拉重量", "荧光", "钻身高度", "桌面", "抛光度", "对称度","腰棱","底尖","量度尺寸","编号"]
        data.to_csv(".\1.csv", encoding='utf_8_sig', index=False, columns=columns)
 
 
    def parseDetail(self, url):
        response = requests.get(url, headers = self.headers)
        detail = {}
        count = 1
        if response.status_code == 200:
            try:
                
                soup = BeautifulSoup(response.text, "html.parser")
                detail["序号"] = count
                count = count + 1
                detail["价格"] = soup.find("span", class_ = "price-label").text
                detail["成色"] = soup.find("div", class_ = "color-column-1").text
                detail["净度"] = soup.find("div", class_ = "clarity-column-1").text
                detail["克拉重量"] = soup.find("div", class_ = "carat-weight-column-1").text
                detail["荧光"] = soup.find("div", class_ = "fluorescence-column-1").text
                detail["钻身高度"] = soup.find("div", class_ = "depth--column-1").text
                detail["桌面"] = soup.find("div", class_ = "table--column-1").text
                detail["抛光度"] = soup.find("div", class_ = "polish-column-1").text
                detail["对称度"] = soup.find("div", class_ = "symmetry-column-1").text
                detail["腰棱"] = soup.find("div", class_ = "girdle-column-1").text
                detail["底尖"] = soup.find("div", class_ = "culet-column-1").text
                detail["量度尺寸"] = soup.find("div", class_ = "measurements-column-1").text
                detail["编号"] = soup.find("div", class_ = "stock-number-column-1").text
            
                return detail
            except:
                return "wrong1"
        else:
            print("wrong2")
 
if __name__ == "__main__":
    Bluenile = bluenileSpider()
    Bluenile.parsePage(URL1)
```
