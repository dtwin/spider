```python
# -*- coding:UTF-8 -*-
from bs4 import BeautifulSoup
import requests, sys
from contextlib import closing
    
class get_photos(object):

    def __init__(self):
        self.photos_id = []
        self.download_server = 'pic.onlylady.com/files/download.php?id=13332195'
        self.target = 'http://pic.onlylady.com/2017/0614/3904680.shtml'
        
    def get_ids(self):
            req = requests.get(url=self.target)
            req.encoding = 'gbk'
            html = req.text
            div_bf = BeautifulSoup(html)
            bf =div_bf.find_all('a',class_ = "download")
            
            
            
        
    def download(self, photo_id, filename):
            
            #target = self.download_server.replace('xxx', photo_id)
            with closing(requests.get(url=photo_id)) as r:
                with open('%d.jpg' % filename, 'ab+') as f:
                    for chunk in r.iter_content(chunk_size = 1024):
                            if chunk:
                                f.write(chunk)
                                f.flush()
if __name__ == '__main__':
    gp = get_photos()
    print('获取图片连接中:')
    gp.get_ids()
    print('图片下载中:')
    
    print(' 正在下载第1张图片')
    for i in range(1):
        gp.download(bf,i+1)
        
    
    
   
    
    
    
    
    
    
```
